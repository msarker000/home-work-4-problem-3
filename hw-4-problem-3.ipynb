```python
import pandas as pd
data = pd.read_csv('titanic.csv')
data
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Plot histogam for passenger Survived
plt.hist(data['Survived'], bins=10)
plt.title('Histogram of passenger Survived')
plt.show()
```

```python
#Plot histogam for passenger class
plt.hist(data['Pclass'], bins=10)
plt.title('Histogram of passenger class')
plt.show()
```

```python
#Plot histogam for passenger age
plt.hist(data['Age'].dropna(), bins=10)
plt.title('Histogram of passenger age')
plt.show()
```

```python
#Plot histogam for passenger parch
plt.hist(data['Parch'].dropna(), bins=10)
plt.title('Histogram of passenger Parch')
plt.show()
```

```python
#Plot histogam for passenger fare
plt.hist(data['Fare'].dropna(), bins=50)
plt.title('Histogram of passenger fare')
plt.show()
```

```python
# Plot  bar charts for passenger Survived
data['Survived'].value_counts().plot(kind='bar')
plt.title('Bar chart of Survived')
plt.show()
```

```python
# Plot  bar charts for passenger Embarked
data['Embarked'].value_counts().plot(kind='bar')
plt.title('Bar chart of Embarked')
plt.show()
```

```python
# Plot  bar charts for passenger Sex
data['Sex'].value_counts().plot(kind='bar')
plt.title('Bar chart of Sex')
plt.show()
```

```python
# Plot  bar charts for passenger SibSp
data['SibSp'].value_counts().plot(kind='bar')
plt.title('Bar chart of SibSp')
plt.show()
```

```python
# Plot  bar charts for passenger Cabin
data['Cabin'].value_counts().plot(kind='bar')
plt.title('Bar chart of Cabin')
plt.show()
```

```python
data.groupby(['Sex','Survived'])['Name'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
print('as showing graph, the ration of female servived is gratter than male' )
print('for female the ratio of servived is higher itself')
print('for male Ratio of not servived is higher itself')
```

```python
data.groupby(['Sex','Pclass'])['Name'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
print('In Both case, Ratio of same for both male and Female')
```

```python
data.groupby(['Survived','Pclass'])['Name'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
print('Here Import information, In Not Survived group, more number of class-3 passenger and class-1 and class- 2  are same')
print('In Survived group, More numver of class-1 and class-2 people')
```
